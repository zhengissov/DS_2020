```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

img1 = cv2.imread('/content/photo.jpg',cv2.IMREAD_GRAYSCALE)
img_resized1 = cv2.resize(img1,(500, 500))
cv2_imshow(img_resized1)
```

```
(h, w) = img_resized1.shape[:2]
center = (w / 2, h / 2)
scale = 1.0

M = cv.getRotationMatrix2D(center, 180, scale)
rotated270 = cv.warpAffine(img_resized1, M, (w, h))

plt.imshow(rotated270, 'gray')
```
